```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes físicas
h = 6.62607015e-34  # Constante de Planck (J·s)
c = 2.99792458e8    # Velocidade da luz (m/s)
k = 1.380649e-23    # Constante de Boltzmann (J/K)

# Temperatura do planeta WASP-96 b
T = 709  # Kelvin

# Comprimentos de onda (de UV até infravermelho médio)
wavelength = np.linspace(0.2e-6, 25e-6, 500)  # metros

# Lei de Planck - radiância espectral
def planck(wl, T):
    return (2*h*c**2) / (wl**5 * (np.exp((h*c)/(wl*k*T)) - 1))

# Cálculo da radiação
radiance = planck(wavelength, T)

# Lei de Wien - pico da emissão
lambda_peak = 2.898e-3 / T  # metros

# Plot
plt.figure(figsize=(10,6))

# Região do espectro visível
plt.axvspan(0.4, 0.7, alpha=0.15, color='gray', linestyle='--',
            edgecolor='gray', linewidth=1.5, label='Espectro visível')
plt.axvline(0.4, color='gray', linestyle='--', linewidth=1, alpha=0.5)
plt.axvline(0.7, color='gray', linestyle='--', linewidth=1, alpha=0.5)

# Curva de Planck
plt.plot(wavelength*1e6, radiance, linewidth=2, label='Radiação de corpo negro')

# Marcação do pico
plt.scatter(lambda_peak*1e6, planck(lambda_peak, T), marker='o', color='C0', s=80, zorder=5)
plt.text(lambda_peak*1e6 + 0.3, planck(lambda_peak, T)*1.0,
         f"Pico ≈ {lambda_peak*1e6:.2f} µm", fontsize=12)

# Configurações do gráfico
plt.xlabel("Comprimento de onda (µm)")
plt.ylabel("Radiância espectral")
plt.title("Curva de Corpo Negro — WASP-96 b (T = 709 K)")
plt.grid(True, alpha=0.4)

plt.legend()
plt.tight_layout()
plt.show()
```
